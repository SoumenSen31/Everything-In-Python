# Counter - Count Elements

```python
from collections import Counter

my_list = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
count = Counter(my_list)
print(count)
```

# Food List Counter

```python
foods = ["banana", "apple", "sushi", "ritu", "apple"]
counts = Counter(foods)
print(counts)
print("\nTop 2:", counts.most_common(2))
```

```python
import matplotlib.pyplot as plt
from collections import Counter
grades = [83,95,91,87,70,0,85,82,100,63,73,77,0]
histogram = Counter(min(grade//10*10,90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor="red",color="green")
plt.axis([-5,105,0,5])
plt.xticks([10 * i for i in range(11)])
plt.show()
```

# LINE CHARTS

```python
import matplotlib.pyplot as plt
variance = [1,2,4,8,16,32,64,128,256]
bias_squared = [256,128,64,32,16,8,4,2,1]
total_error = [x+y for x ,y in zip(variance,bias_squared)]
xs = [i for i,_ in enumerate(variance)]
plt.plot(xs,variance,'g-',label = 'varience')
plt.plot(xs,bias_squared,'r-',label = "bias^2")
plt.plot(xs,total_error,'b:',label = "total_error")

plt.legend(loc=9)
plt.title("the bias-varience tradeoff")
plt.xticks([])
plt.xlabel("model complexity")
plt.show()
```

# scatterplots

```python
import matplotlib.pyplot as plt
friends = [70,65,72,63,71,64,60,64,67]
minutes = [175,170,205,120,220,130,105,145,190]
labels = ['a','b','c','d','e','f','g','h','i']

plt.scatter(friends,minutes)

for label,friends_count,minute_count in zip(labels,friends,minutes):
    plt.annotate(label,
                 xy = (friends_count,minute_count),
                 xytext = (5,-5),
                 textcoords='offset points')
plt.show()
```
