# Nominal GDP

```python
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9,2862.5,5975.6,10289.7,14958.3]

plt.plot(years,gdp,color="green",marker="o",linestyle="solid")
plt.title("Normal GDP")
plt.xlabel("year")
plt.ylabel("Billions of $")
plt.show()
```

# Counter - Count Elements

```python
from collections import Counter

my_list = ['apple', 'banana', 'apple', 'orange', 'banana', 'banana']
count = Counter(my_list)
print(count)
```

```python
count_chars = Counter("mississippi")
print(count_chars)
```

# Food List Counter

```python
foods = ["banana", "apple", "sushi", "ritu", "apple"]
counts = Counter(foods)
print(counts)
print("\nTop 2:", counts.most_common(2))
```

```python
grades = [83,95,91,87,70,0,85,82,100,63,73,77,0]
histogram = Counter(min(grade//10*10,90) for grade in grades)
plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor=(0,0,0))
plt.axis([-5,105,0,5])
plt.xticks([10 * i for i in range(11)])
plt.show()
```
