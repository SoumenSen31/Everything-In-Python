```python
# Score Range  | Frequency
# 0-20         | 2 students
# 21-40        | 3 students
# 41-60        | 15 students
# 61-80        | 25 students
# 81-100       | 5 students
import matplotlib.pyplot as plt

score_range = ["0-20","21-40","41-60","61-80","81-100"]
frequency = [2,3,15,25,5]
plt.bar(score_range,frequency)
plt.title("Histogram")
plt.xlabel("score-range")
plt.ylabel("frequency")
plt.show()
```

Nominal GDP

```python
import matplotlib.pyplot as plt
years = [1950,1960,1970,1980,1990,2000,2010]
gdp = [300.2,543.3,1075.9,2862.5,5975.6,10289.7,14958.3]

plt.plot(years,gdp,color="green",marker="o",linestyle = "solid")

plt.title("Normal GDP")
plt.xlabel("year")
plt.ylabel("Billions of $")
plt.show()
```
