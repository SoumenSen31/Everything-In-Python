```python
# Score Range  | Frequency
# 0-20         | 2 students
# 21-40        | 3 students
# 41-60        | 15 students
# 61-80        | 25 students
# 81-100       | 5 students
import matplotlib.pyplot as plt

score_range = ["0-20","21-40","41-60","61-80","81-100"]
frequency = [2,3,15,25,5]
plt.bar(score_range,frequency)
plt.title("Histogram")
plt.xlabel("score-range")
plt.ylabel("frequency")
plt.show()
```
